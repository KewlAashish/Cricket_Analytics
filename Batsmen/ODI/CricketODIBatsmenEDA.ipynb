```python
import re
def selectingreigon(player):
  if re.search('\(.*', player):
  
        # Extract the position of beginning of pattern
        pos = re.search('\(.*', player).start()
  
        # return the cleaned name
        return player[pos+1:-1]
  
  else:
        # if clean up needed return the same name
        return player

  return player
```

```python
df['Region'] = df['Player'].apply(selectingreigon)
```

```python
df.head(10)
```

```python
def Clean_names(City_name):
    # Search for opening bracket in the name followed by
    # any characters repeated any number of times
    if re.search('\(.*', City_name):
  
        # Extract the position of beginning of pattern
        pos = re.search('\(.*', City_name).start()
  
        # return the cleaned name
        return City_name[:pos]
  
    else:
        # if clean up needed return the same name
        return City_name
```

```python
df['Player'] = df['Player'].apply(Clean_names)
```

CHANGING THE SPAN COLUMN BY KEEPING ONLY THE YEARS THEY'VE PLAYED

```python
def time(column):
  years = int(column[0:4]) - int(column[5:])
  return -1*years
```

CLEANING THE REGION COLUMN BY JUST KEEPING THE COUNTRY THEY PLAYED FOR AND NOT THE OTHER PART

```python
df['Region'].unique()
```

```python
def cleaning_region(col):
  if col == 'INDIA' or col == 'Asia/INDIA' or col == 'Asia/ICC/INDIA' :
    col = 'IND'
  elif col == 'Asia/ICC/SL' or col == 'SL' or col == 'Asia/SL':
    col = 'SL'
  elif col == 'Afr/ICC/SA' or col == 'Afr/SA' or col == 'ICC/SA' or col == 'NL/SA' or col == 'SA/USA' or col == 'SA' or col == 'AUS/SA' :
    col = 'SA'
  elif col == 'AUS/ICC' or col == 'AUS':
    col = 'AUS'
  elif col == 'Asia/PAK' or col == 'PAK' or col == 'Asia/ICC/PAK' or col == '3) (PAK':
    col = 'PAK'
  elif col == 'ICC/WI' or col == 'USA/WI' or col == 'WI':
    col = 'WI'  
  elif col == 'CAN/WI' or col == 'CAN':
    col = 'CAN'
  elif col == 'NZ' or col == 'HKG/NZ' or col == 'AUS/NZ' or col == 'ICC/NZ':
    col = 'NZ'
  elif col == 'BDESH' or col == 'Asia/BDESH':
    col = 'BDESH'
  elif col == 'ENG/SCOT':
    col = 'SCOT'
  elif col == 'ENG/IRE' or col == 'IRE':
    col = 'IRE'
  elif col == 'EAf' or col == 'KENYA' or col == 'Afr/KENYA':
    col = 'KEN'
  elif col == '1) (UAE' or col == 'UAE':
    col = 'UAE'
  elif col == 'ENG/PNG' or col == 'ENG' or col == 'ENG/ICC':
    col = 'ENG'
  elif col == 'Afr/ZIM':
    col = 'ZIM'
  return col
```

```python
df['Region'] = df['Region'].apply(cleaning_region)
```

CLEANING THE HS COLUMN BECAUSE OUT OR NOT OUT DOESN'T MATTER FOR PREDICTION WE'RE TRYING TO MAKE

```python
def cleanhs(hs):
  if hs[-1] == '*':
    return float(hs[:-1])
  elif '-' in hs:
    return 0
  else:
    return float(hs)
```

TO FIND THE TOP 10 BEST BATSMEN FOR ODI OF ALL TIME, SOME PERFORMANCE METRICS SHOULD BE CREATED.

THE PERFORMANCE CRITERIA I'VE KEPT IS:

1. BETTER AVERAGE: Average matters a lot for other people to judge how good a batsman actually is, if batting average isn't that good but he has a great strike rate, it doesn't matter.
2. NUMBER OF HUNDREDS : Number of hundreds scored should have two times the importance of the number of fifties scored by the batsman.
3. NUMBER OF FIFTIES : More the fifties better the batsman is.
4. LESS NUMBER OF DUCKS: Ducks shouldn't be appreciated (negative)
5. MORE RUNS : Runs are great but they should be divided by the number of innings played by the batsman.
6. MORE BALLS FACED : This is also a great factor of evaluation but they should be divided by the number of innings played by the batsman.
7. MORE NOT OUTS(NO) : They are a great factor for evaluation too but not much important.
8. GOOD SR : Better the strike rate better the player right divided by the number of innings played by him.

score = number_hundred*2 + number_fifties - number_ducks + average*number_innings + (runs + 0.5*balls + 0.5*NO)/number_innings 

```python
#DROPPING COLUMNS WHERE NUMBER OF INNINGS IS ZERO:
df = df[df['Inns'] != 0]
```

```python
df['score'] = df['Hundreds']*2/df['Hundreds'].mean() + df['Fifties']/df['Fifties'].mean() - df['Ducks']/df['Ducks'].mean() + (df['Ave']/df['Ave'].mean())*(df['Inns']/df['Inns'].mean()) + (df['Runs']/df['Runs'].mean() + 0.75*df['BF']/df['BF'].mean() + 0.5*df['NO']/df['NO'].mean())/(df['Inns']/df['Inns'].mean())
```
