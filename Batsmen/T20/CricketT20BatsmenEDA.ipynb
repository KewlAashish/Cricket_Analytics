```python
df.head()
```

CREATING REGION COLUMN FROM THE VALUES IN PLAYER COLUMN WHICH ARE INSIDE THE BRACKET AND FOR THAT I'LL BE CREATING A FUNCTION.

```python
import re
def selectingreigon(player):
  if re.search('\(.*', player):
  
        # Extract the position of beginning of pattern
        pos = re.search('\(.*', player).start()
  
        # return the cleaned name
        return player[pos+1:-1]
  
  else:
        # if clean up needed return the same name
        return player

  return player
```

```python
df['Region'] = df['Player'].apply(selectingreigon)
```

```python
df.head(10)
```

SINCE I HAVE CREATED A COLUMN FOR REGION FROM WHICH THE PLAYERS BELONG, THE NEXT THING IS TO REMOVE THAT REGION PART FROM THE PLAYER'S NAME COLUMNS. I'VE CREATED A FUNCTION FOR THAT AND APPLIED IT ON THE PLAYER COLUMN.

```python
def Clean_names(City_name):
    # Search for opening bracket in the name followed by
    # any characters repeated any number of times
    if re.search('\(.*', City_name):
  
        # Extract the position of beginning of pattern
        pos = re.search('\(.*', City_name).start()
  
        # return the cleaned name
        return City_name[:pos]
  
    else:
        # if clean up needed return the same name
        return City_name
```

```python
df['Player'] = df['Player'].apply(Clean_names)
```

```python
df.head()
```

CHANGING THE SPAN COLUMN TO THE YEARS PLAYER PLAYED BY COMPUTING THE DIFFERENCE BETWEEN THE YEARS.

```python
def time(column):
  years = int(column[0:4]) - int(column[5:])
  return -1*years
```

```python
df['Span'] = df['Span'].apply(time)
```

TIME TO CLEAN THE REGION COLUMN TO ONLY KEEP THE COUNTRY THEY'VE PLAYED FOR AND FOR THAT I'VE CREATED A FUNCTION

```python
def cleaning_region(col):
  if col == 'INDIA' or col == 'ICC/INDIA' :
    col = 'IND'
  elif col == 'ICC/SL/World' or col == 'SL' :
    col = 'SL'
  elif col == 'SA/World' or col == 'NL/SA' or col == 'SA' :
    col = 'SA'
  elif col == 'AUS/World' or col == 'AUS' or col == 'AUS/NL':
    col = 'AUS'
  elif col == '1) (PAK' or col == 'PAK' or col == 'ICC/PAK' or col == '3) (PAK' or col == '2) (PAK':
    col = 'PAK'
  elif col == 'WI/World' or col == 'USA/WI' or col == 'WI':
    col = 'WI'  
  elif col == 'ICC/NEPAL':
    col = 'NPL'
  elif col == 'NZ' or col == 'AUS/ICC/NZ' or col == 'HKG/NZ' or col == 'ICC/NZ' or col == 'NZ/World'  :
    col = 'NZ'
  elif col == 'BDESH' or col == 'BDESH/ICC/World':
    col = 'BDESH'
  elif col == 'ENG/IRE' or col == 'IRE':
    col = 'IRE'
  elif col == 'ENG/World' or col == 'ENG' or col == 'ENG/ICC':
    col = 'ENG'
  elif col == 'AFG/ICC' or col == 'AFG/GER':
    col = 'AFG'
  elif col == '2) (DEN':
    col = 'DEN'
  return col
```

```python
df['Region'] = df['Region'].apply(cleaning_region)
```

CLEANING THE COLUMN's BY REMOVING THE * AND - (HYPHEN) IN THE SCORE BY CREATING A FUNCTION FOR THAT AND APPLY IT ON THE HS COLUMN

```python
def cleanhs(hs):
  if hs[-1] == '*':
    return float(hs[:-1])
  elif '-' in hs:
    return 0
  else:
    return float(hs)
```

```python
df['HS'] = df['HS'].apply(cleanhs)
```

```python
df['Inns'] = df['Inns'].apply(cleanhs)
```

```python
df['NO'] = df['NO'].apply(cleanhs)
```

```python
df['Runs'] = df['Runs'].apply(cleanhs)
```

```python
df['Ave'] = df['Ave'].apply(cleanhs) #Ave , BF , SR , Hundreds , Fifties , Ducks
```

```python
df['BF'] = df['BF'].apply(cleanhs)
```

```python
df['SR'] = df['SR'].apply(cleanhs)
```

```python
df['Hundreds'] = df['Hundreds'].apply(cleanhs)
```

```python
df['Fifties'] = df['Fifties'].apply(cleanhs)
```

```python
df['Ducks'] = df['Ducks'].apply(cleanhs)
```

```python
df['Fours'] = df['Fours'].apply(cleanhs)
```

```python
df['Sixes'] = df['Sixes'].apply(cleanhs)
```
